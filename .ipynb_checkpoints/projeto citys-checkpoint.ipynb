```python
import pandas as pd
import numpy as np

import datetime as dt 
import time
```

```python
citys = pd.read_csv('city_data.csv',sep='|',header=1, index_col="City")
citys
```

```python
citys.head()
```

```python
citys.isna().sum() #há missing values, temos que os retirar e trocar por outros que vamos pesquisar
```

```python
#retirar coluna Average Price Groceries. (feito)
#pesquisar os outros missing values em falta (feito)
#mudar os nomes das colunas
```

```python
citys.columns #confirmar nomes das colunas do nosso dataset
```

```python
citys1 = citys.drop(columns=['Average Price Groceries'],axis=1) #retiramos esta coluna pois esta coluna tem 80 missing values sendo que existem 86 linhas achando assim que não é uma variavel importante.
citys1
```

```python
citys1[citys1['Population Density'].isnull()]# na coluna population Density, a cidade de valencia apresenta missing value. 
#pesquisamos na wikipedia e a densidade populacional é de 5 865.2 . Alterar o missing value da coluna para este novo valor
```

```python
citys1['Population Density'] = citys1['Population Density'].fillna(5865.2)#através da função fillna preenchemos o missing value existente na coluna population density com um valor retirado da wikipedia
citys1[citys1['Population Density'].isnull()]#para confirmar que o missing value daquela coluna desapareceu
```

```python
citys1[citys1["Working Age Population "].isnull()]#na coluna working age population, a cidade de Dusseldorf apresenta missing value
#https://www.arbeitsagentur.de/vor-ort/duesseldorf/statistik? - retiramos daqui o valor em falta
```

```python
citys1['Working Age Population '] = citys1['Working Age Population '].fillna(424159.0)
citys1[citys1['Working Age Population '].isnull()]
```

```python
citys1[citys1["Unemployment Rate"].isnull()] #nesta coluna da taxa de desemprego, a cidade de Gent apresenta missing values
#https://provincies.incijfers.be/databank/report/?id=rapport_economie&input_geo=gemeente_44021%2C&utm- retiramos desta tabela o valor de 10.8% da taxa de desemprego no ano de 2024
```

```python
citys1["Unemployment Rate"]=citys1["Unemployment Rate"].fillna(10.8)
citys1[citys1["Unemployment Rate"].isnull()]
```

```python
citys1[citys1["GDP per Capita"].isnull()]#nesta coluna a cidade de Lemesos (Cyprus) tem missing values.
#https://metroverse.hks.harvard.edu/city/4065/economic-composition?utm- retiramos desta tabela o valor de 4.170 dolares de gdp per capita desta cidade e convertemos em euros (3598.40)
```

```python
citys1["GDP per Capita"]=citys1["GDP per Capita"].fillna(3598.40)
citys1[citys1["GDP per Capita"].isnull()]
```

```python
citys1[citys1["Main Spoken Languages"].isnull()] #nesta coluna a cidade de Dusseldorf tem missing values nas linguas mais faladas
#https://citiesinsider.com/country/germany/dusseldorf/languages/en?ut - retiramos daqui as linguas mais faladas em Dusseldorf que são o alemao e o ingles
```

```python
citys1["Main Spoken Languages"]=citys1["Main Spoken Languages"].fillna("German,English")
citys1[citys1["Main Spoken Languages"].isnull()]
```
